```python
import numpy as np
import random
```

```python
class Game:
    take3Point = False

    def reset(self):
        self.pos = (0,2)
        self.state = 6
        self.max_action = 10
        return self.state #state

    def play_step(self, action):
        self.max_action -= 1
        if self.max_action < 0: return self.state, -10, True
        done = False
        r = 0

        # nouvelle position
        if action == 0:
            new_pos = (self.pos[0], self.pos[1] - 1)
        elif action == 1:
            new_pos = (self.pos[0] + 1, self.pos[1])
        elif action == 2:
            new_pos = (self.pos[0], self.pos[1] + 1)
        elif action == 3:
            new_pos = (self.pos[0] - 1, self.pos[1])
        
        # cases spÃ©ciales
        if new_pos == (2,0):
            done = True
            r += 1
        elif new_pos == (2,2) and not self.take3Point:
            self.take3Point = True
            r += 10
        elif new_pos == (1,1):
            self.pos = new_pos
            return 4, -1, False
        
        # position en dehors de la map
        if new_pos[0] < 0 or new_pos[0] > 2:
            return self.state, 0, False
        if new_pos[1] < 0 or new_pos[1] > 2:
            return self.state, 0, False
        
        # autre cas
        self.pos = new_pos
        self.state = self.pos[0] + 3 * self.pos[1]

        return self.state, r, done #new_state, recompense, done
```

```python
def make_action(state, Q_table:np.ndarray, eps=0):
    if random.uniform(0,1) < eps:
        return random.randint(0, 3)
    else:
        return np.argmax(Q_table[state])
```

```python
env = Game()

#hyper parametre
n_game = 100
gamma = 0.8
learning_rate = 0.1

MAX_STATE = 9
ACTIONS = 4

Q_table = np.zeros((MAX_STATE, ACTIONS),dtype=float)
```

```python
for _ in range(n_game):
    state = env.reset()
    done = False
    while not done:

        action = make_action(state, Q_table, eps=1)

        new_state, recompense, done = env.play_step(action)

        # Update Q_table
        new_action = make_action(state, Q_table, eps=0)
        Q_table[state][action] += learning_rate*(recompense + gamma*Q_table[new_state][new_action] - Q_table[state][action])

        state = new_state
```
